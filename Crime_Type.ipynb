```python
import os
import pandas as pd

# Define the input file path
input_file_path = r'C:\Users\jckat\OneDrive\Documents\GitHub\Space_TIme_Exploration_of_Metro_Police\cleaned_data.csv'

# Define the output directory
output_dir = r'C:\Users\jckat\OneDrive\Documents\GitHub\Space_TIme_Exploration_of_Metro_Police'

# Create the directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load the dataset
metro_df = pd.read_csv(input_file_path)

# Convert 'Month' column to datetime format, handling errors
metro_df['Month'] = pd.to_datetime(metro_df['Month'], errors='coerce')

# Remove rows with null values in the 'Month' column
metro_df = metro_df.dropna(subset=['Month'])

# Get unique crime types
crime_types = metro_df['Crime type'].unique()

# Create a CSV file for each crime type
for crime in crime_types:
    # Filter the dataframe for the current crime type
    crime_df = metro_df[metro_df['Crime type'] == crime]

    # Create a filename based on the crime type
    filename = f"{crime.replace(' ', '_').lower()}_crimes.csv"

    # Create the full path for the output file
    file_path = os.path.join(output_dir, filename)

    # Save the filtered dataframe to a CSV file
    crime_df.to_csv(file_path, index=False)

print(f"CSV files created for each crime type in the directory: {output_dir}")

```
